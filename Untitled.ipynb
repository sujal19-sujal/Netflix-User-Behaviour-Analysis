```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
shows_df = pd.read_csv('shows.csv')
users_df = pd.read_csv('users.csv')
viewing_history_df = pd.read_csv('viewing_history.csv')

# Display the first few rows and information for each DataFrame
print("Shows DataFrame:")
print(shows_df.head())
print("\nShows DataFrame Info:")
shows_df.info()

print("\nUsers DataFrame:")
print(users_df.head())
print("\nUsers DataFrame Info:")
users_df.info()

print("\nViewing History DataFrame:")
print(viewing_history_df.head())
print("\nViewing History DataFrame Info:")
viewing_history_df.info()
```

```python
# Merge viewing history with show information
merged_df = pd.merge(viewing_history_df, shows_df, on='show_id')

# Merge with user information
full_df = pd.merge(merged_df, users_df, on='user_id')

# Split genres into separate rows to count them individually
full_df['genre'] = full_df['genre'].str.split(', ')
full_df_exploded = full_df.explode('genre')

# Convert 'watch_date' to datetime objects
full_df['watch_date'] = pd.to_datetime(full_df['watch_date'])

print("\nMerged and Processed DataFrame:")
print(full_df_exploded.head())
```

```python
# Count the occurrences of each genre
top_genres = full_df_exploded['genre'].value_counts()

print("\nTop 5 Most-Watched Genres:")
print(top_genres.head())

# Create a bar chart for visualization
plt.figure(figsize=(12, 7))
sns.barplot(x=top_genres.index, y=top_genres.values, palette='coolwarm')
plt.title('Most Watched Genres', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Number of Views', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Group by user and sum the watch duration
watch_hours_per_user = full_df.groupby('username')['watch_duration_minutes'].sum()

# Convert minutes to hours and sort
watch_hours_per_user_hours = (watch_hours_per_user / 60).sort_values(ascending=False)

print("\nTop 5 Users by Watch Hours:")
print(watch_hours_per_user_hours.head())

# Create a bar chart for visualization
plt.figure(figsize=(10, 6))
sns.barplot(x=watch_hours_per_user_hours.index, y=watch_hours_per_user_hours.values, palette='viridis')
plt.title('Watch Hours Per User', fontsize=16)
plt.xlabel('Username', fontsize=12)
plt.ylabel('Total Watch Hours', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Count the number of views per user and show
binge_behavior = full_df.groupby(['username', 'title'])['history_id'].count().reset_index()

# Filter for shows with more than one view
binge_watchers = binge_behavior[binge_behavior['history_id'] > 1]

if not binge_watchers.empty:
    print("\nUsers who Binge-Watched:")
    print(binge_watchers)

    # Create a bar chart for visualization
    plt.figure(figsize=(10, 6))
    sns.barplot(x='title', y='history_id', hue='username', data=binge_watchers, palette='rocket')
    plt.title('Binge-Watching Behavior', fontsize=16)
    plt.xlabel('Show Title', fontsize=12)
    plt.ylabel('Number of Views', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.legend(title='Username')
    plt.tight_layout()
    plt.show()
else:
    print("\nNo binge-watching behavior detected in the dataset (no user watched the same show more than once).")
```
